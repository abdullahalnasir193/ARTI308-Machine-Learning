```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
#load Dataset
df=pd.read_csv("global_cars_enhanced.csv")
df
```

```python
print(df.isna())
```

```python
plt.figure(figsize=(10,5))
df.groupby('Brand')['Price_USD'].mean().sort_values().plot(kind='barh', color='skyblue')
plt.title('Average Price by Brand')
plt.xlabel('Average Price ($)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Horsepower', y='Price_USD', hue='Fuel_Type')
plt.title('Horsepower vs Price')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='Fuel_Type', palette='viridis')
plt.title('Distribution of Fuel Types')
plt.show()
```
